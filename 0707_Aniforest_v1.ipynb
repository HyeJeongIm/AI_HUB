```python
! pip install librosa
```

```python
import librosa

letter_file_path = './samples/%s.mp3' % letter

# 녹음할때 sampling 한다고 합니다.

x, sr = librosa.load(letter_file_path)
print('x', x)
print('sr', sr)
```

```python
# 그래프 확인
import matplotlib.pyplot as plt
plt.plot(x)
```

```python
import IPython
IPython.display.Audio(data = x, rate = sr) # 오디오파일을 불러옵니다
```
