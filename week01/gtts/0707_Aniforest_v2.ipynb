```python
# 녹음할때 sampling 한다고 합니다.

import librosa
import matplotlib.pyplot as plt # 그래프 확인
import numpy as np

string = "안녕하세요"
result = []

for letter in string:

  if letter == ' ':
    x = np.zeros(50000)
  else:

    if not os.path.isfile('./samples/%s.mp3' % letter): # 파일이 다 존재하기 때문에 다운받을 필요 없다
      print('파일 다운로드 저장 및 필요')

      tts = gTTS(letter, lang = lang)
      tts.save('./samples/%s.mp3' % letter) # 폴더명이 자동으로 바뀌도록 
    else:
      letter_file_path = './samples/%s.mp3' % letter
      x, sr = librosa.load(letter_file_path)

  result = np.append(result, x)

plt.plot(result)
```
